Projeto - Redução de dimensionalidade de imagens

Redução de dimensionalidade é utilizada para diminuir a complexidade de uma imagem para usá-la em redes neurais, por exemplo: Uma imagem colorida de 244 x 244 tem 150.528 valores, quando essa mesma imagem passa pelo processo de  grayscale fica com 50.176 valores, em suma, isso diminui a complexidade, e retira informações redundantes, melhorando o treinamento.

```python
# Transformar a imagem para RGB
from PIL import Image #essa  biblioteca vai ser usada somente para fazer a tranformação
                      # de por exemplo um jpg para RGB
import matplotlib.pyplot as plt #Usado para plotar as imagens
#link imagem usada : https://pixabay.com/pt/photos/tulipas-jardim-flores-b%C3%AAn%C3%A7%C3%A3o-9536635/
#So colocar dentro da pasta content do colab
img = Image.open('tulips-9536635_1280.jpg')

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()

img = img.convert('RGB') #Converte a imagem de jpeg para RGB
```

```python
def rgb_to_grayscale(R,G,B): #Transforma para grayscale
  return int(0.299 * R + 0.587 * G + 0.114 * B)

def binarizarPixel(pixel,limiar): # Do Grayscale transforma par binario
  if pixel > limiar:
    return 255
  else:
    return 0

largura,altura  = img.size #Pega as dimensoes da imagem, tanto para:
                           #1 - criar as novas imagens
                           #2 - percorrer a imagem tranformando para grayscale e binario

imagem_cinza = Image.new('L',(largura,altura)) # nova imagem, so usado para poder plotar
imagem_binarizada = Image.new('L',(largura,altura))

for i in range(altura):# Usado para percorrer a imagem

  for j in range(largura):
    r,g,b = img.getpixel((j,i)) #Usa a função para pegar o rgb do pixel
    pixel_cinza = rgb_to_grayscale(r,g,b)#Usa função para  grayscale
    pixel_binarizado = binarizarPixel(pixel_cinza,127) #Usa função para binarizar usando o pixel cinza

    imagem_cinza.putpixel((j,i),pixel_cinza) #Colocas os pixels nas imagens novas
    imagem_binarizada.putpixel((j,i),pixel_binarizado)



```

```python
plt.imshow(imagem_cinza, cmap='gray') #Plot imagem em grayscale
plt.axis('off')
plt.show()
```

```python
plt.imshow(imagem_binarizada, cmap='gray') # plot imagem binarizada
plt.axis('off')
plt.show()
```
